## Importing the **DATASET**

```python

from sklearn import datasets
df_=datasets.load_iris()
# df.target
```

```python
import pandas as pd
df=df_.data
df=pd.DataFrame(df)
# df_.target

```

```python
df
```

# Checking for the optimal value of **K**

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df)
    sse.append(km.inertia_)

plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```

# Training the model

```python

km=KMeans(n_clusters=3)
clus=km.fit_predict(df)
```

```python
print(clus)
```

# Concatinating the predicted values with the DataFrame

```python
df['clus']=clus
```

```python
df
```

# Graphical Representation

```python
df1=df[df.clus==0]
df2=df[df.clus==1]
df3=df[df.clus==2]
```

```python

from matplotlib import pyplot as plt
plt.scatter(df1[0],df1[1],color='green')
plt.scatter(df2[0],df2[1],color='red')
plt.scatter(df3[0],df3[1],color='blue')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.legend()
```
